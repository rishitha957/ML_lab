```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import minimize
```

# Section-A

```python
x = np.linspace(0,20,100)
y = 2.1 * x*x - 49.1 * x + 147.9

plt.xlim([ 0,30 ])
plt.ylim([ -150, 150 ])

plt.plot(x,y,'b.')
```

```python
f = lambda w: w[0]*x*x + w[1]*x + w[2]
error = lambda w : np.sum( (f(w) - y)**2 )/x.shape[0]
```

```python
def grad1 (w) :                       #with normalisation
    tmp = np.zeros(3)
    a = 2*(f(w) - y)/x.shape[0]
    tmp[0] = np.sum(a*x*x)
    tmp[1] = np.sum(a*x)
    tmp[2] = np.sum(a)
    t = np.linalg.norm(tmp)
    tmp = tmp/(t) 

    return tmp
```

```python
def grad2 (w) :                       #without normalisation
    tmp = np.zeros(3)
    a = 2*(f(w) - y)/x.shape[0]
    tmp[0] = np.sum(a*x*x)
    tmp[1] = np.sum(a*x)
    tmp[2] = np.sum(a)
    t = np.linalg.norm(tmp)

    return tmp
```

```python
w_init = np.array([0,0,0])
res = minimize(fun=error,x0=w_init,jac=grad1)     #using grad function with normalisation
print(res.x)
```

```python
res = minimize(fun=error,x0=w_init,jac=grad2)     #using grad function without normalisation
print(res.x)
```

```python
alpha = 0.1      #Experiment for 0.01, 1.0, 10.0, 100.0 and 1000.0
n_iter = 1000000 #Experiment for 100, 1000, 10000, 100000 and 1000000
```

```python

w_old = w_init
for i in range(n_iter) :
  
    w_new = w_old - alpha*grad1(w_old)            #Using grad with normalisation and alpha = 0.1 and n_iter = 1000000
    w_old = w_new + 0

print (w_old)
plt.plot(x,f(w_old))
```

```python
x = np.linspace(0,2*math.pi,100)
y = np.sin(x)


plt.plot(x,y)

f = lambda w: w[0]*x**6 + w[1]*x**5 + w[2]*x**4 +w[3]*x**3 + w[4]*x**2 + w[5]*x +w[6]
error = lambda w : np.sum( (f(w) - y)**2 )
```

```python
def grad (w) :                       #with normalisation
    tmp = np.zeros(7)
    a = 2*(f(w) - y)
    tmp[0] = np.sum(a*x**6)
    tmp[1] = np.sum(a*x**5)
    tmp[2] = np.sum(a*x**4)
    tmp[3] = np.sum(a*x**3)
    tmp[4] = np.sum(a*x**2)
    tmp[5] = np.sum(a*x)
    tmp[6] = np.sum(a)
    
    t = np.linalg.norm(tmp)
    tmp = tmp/(t) 

    return tmp
```

```python
w_init = np.array([0,0,0,0,0,0,0])
w_old = w_init
alpha = 0.1
n_iter = 1000000
for i in range(n_iter) :
  
    w_new = w_old - alpha*grad(w_old)            #Using grad with normalisation and alpha = 0.01 and n_iter = 1000000
    w_old = w_new + 0

print (w_old)
```

```python
w_init = np.array([0,0,0,0,0,0,0])
res = minimize(fun=error,x0=w_init,jac=grad)     #using grad function with normalisation
print(res.x)
```
